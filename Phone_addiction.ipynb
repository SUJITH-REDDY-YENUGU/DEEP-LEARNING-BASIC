```python
# #CONVERTING OBJECTS TO THE INTEGER FORMAT
from sklearn.preprocessing import LabelEncoder
Encoder_objects={}
before_objects=[]
for i in object_columns:
        encoder=LabelEncoder()
        before_objects.append(list(data_frame[i].unique()))
        data_frame[i]=encoder.fit_transform(data_frame[i])
        Encoder_objects[encoder]=encoder.classes_
```

```python
import math

print("finally the dataframe is free of objects values ....")
print(f"converting the dataframe to numpy so that it is helpful to load  to pytorch")
print(columns_of_df:=data_frame.columns)
data_frame["Addiction_Level"]=data_frame["Addiction_Level"].apply(lambda x :math.ceil(x))
print(data_frame["Addiction_Level"].value_counts())
```

```python
df_features=data_frame.drop(columns=["Addiction_Level"],inplace=False)
df_labels=data_frame["Addiction_Level"]
print(df_features.shape)
print(df_labels.shape)
print(df_labels)
```

```python
import torch as t
from torch.utils.data import DataLoader
from torch.utils.data import Dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_features,df_labels,train_size=0.8,random_state=42)
```

```python
y_train=y_train.apply(lambda y:y-1)
y_test=y_test.apply(lambda y:y-1)       

list_of_train_test=[x_train,x_test,y_train,y_test]

for i in list_of_train_test:
        print(i.shape)

x_train=x_train.to_numpy()
x_test=x_test.to_numpy()
y_train=y_train.to_numpy()
y_test=y_test.to_numpy()
for i in list_of_train_test:
        print(type(i))
```

```python
print(type(x_train))
print(type(x_test))
print(type(y_train))
print(type(y_test))

x_tensor_train=t.from_numpy(x_train)
x_tensor_test=t.from_numpy(x_test)
y_tensor_train=t.from_numpy(y_train)
y_tensor_test=t.from_numpy(y_test)
```

```python
list_of_tensor_train_test=[x_tensor_train,x_tensor_test,y_tensor_train,y_tensor_test]
for i in list_of_tensor_train_test:
        print(i.shape)
```

```python
class Dataset_class(Dataset):
        def __init__(self,x,y):
                self.x_dataset=x
                self.y_dataset=y
        def __len__(self):

                return len(self.y_dataset)
        def __getitem__(self, index):
                return self.x_dataset[index,:],self.y_dataset[index]
        
```

```python
train_dataset=Dataset_class(x_tensor_train,y_tensor_train)
test_dataset=Dataset_class(x_tensor_test,y_tensor_test)
```

```python
train_data_loader=DataLoader(
        dataset=train_dataset,
        shuffle=True,
        batch_size=32,
        drop_last=True,
        num_workers=0

)
test_data_loader=DataLoader(
        dataset=test_dataset,
        shuffle=True,
        batch_size=32,
        drop_last=True,
        num_workers=0
)

```

```python
#writing the model 
import torch.nn as nn
class Model_for_classification(t.nn.Module):
        def __init__(self,input_size,output_size,hidden_size):
                super().__init__()
                self.Layer1=t.nn.Sequential(
                        nn.Linear(input_size,hidden_size),
                        nn.ReLU(),
                        nn.Linear(hidden_size,hidden_size),
                        nn.Dropout(p=0.2),
                        nn.BatchNorm1d(num_features=hidden_size),
                        nn.Linear(hidden_size,hidden_size),
                        nn.Linear(hidden_size,hidden_size*2),
                        nn.Dropout(p=0.2),
                        nn.Linear(hidden_size*2,output_size)
                        
                )
        def forward(self,x):
                return self.Layer1(x)
```

```python
Model=Model_for_classification(input_size=21,output_size=10,hidden_size=64)
```

```python
from torchinfo import summary
summary(
        model=Model,
        input_size=(32,21),
        col_names=["input_size","output_size","num_params","trainable"]
)
```

```python
optimizer=t.optim.Adam(params=Model.parameters(),lr=1e-3)
lossfn=t.nn.CrossEntropyLoss()
```

```python
#training the model 
train_loss_list=[]
test_loss_list=[]
train_acc_list=[]
test_acc_list=[]
train_loss=0
test_loss=0
train_acc=0
test_acc=0
Model.to(t.float32)
epochs=100
for epoch in range(epochs):
        Model.train()
        train_loss=0
        test_loss=0
        train_acc=0
        test_acc=0
        for (X,y) in train_data_loader:
                X=X.to(t.float32)
                y_preds=Model(X)
                loss=lossfn(y_preds,y)
                train_loss+=(loss.item()*32)
                train_acc+=(((y_preds.argmax(dim=1))==y).sum().item())
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
        Model.eval()
        with t.inference_mode():
                for (X,y) in test_data_loader:
                        X=X.to(t.float32)
                        y_preds=Model(X)
                        loss=lossfn(y_preds,y)
                        test_loss+=(loss.item()*32)
                        test_acc+=((y_preds.argmax(dim=1)==y).sum().item())
        train_acc_list.append(train_acc/len(train_data_loader.dataset))
        test_acc_list.append(test_acc/len(test_data_loader.dataset))
        train_loss_list.append(train_loss/len(train_data_loader.dataset))
        test_loss_list.append(test_loss/len(test_data_loader.dataset))
        print(f"\033[1;36m {"="*140}\033[0m")
        print(f"\033[1;36m Epoch : {epoch}\033[0m")
        print(f"train_accuracy : {train_acc_list[epoch]}")
        print(f"test_accuracy : {test_acc_list[epoch]}")
        print(f"train_loss : {train_loss_list[epoch]}")
        print(f"test_loss : {test_loss_list[epoch]}")
```
